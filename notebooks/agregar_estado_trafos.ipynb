# Actualización de Metadatos de Transformadores (Lógica Simple)

Este notebook lee el archivo existente `metadatos_ATRs.csv` y le agrega la columna `En_Servicio`.
**Regla Simple:** Si el código del trafo termina en "4" o "8", es Reserva (NO). Todo lo demás es Operativo (SI).

```python
import pandas as pd
from pathlib import Path
```

```python
# 1. Cargar el CSV existente
ruta_csv = Path("../data/metadatos_ATRs.csv")
df_trafos = pd.read_csv(ruta_csv)

# 2. Lógica Simple para identificar Reservas
def identificar_reserva(row):
    codigo = str(row["Codigo_Trafo"]).strip()
    
    # Regla Global: Si termina en 4 o 8, es reserva
    if codigo.endswith("4") or codigo.endswith("8"):
        return "NO"
            
    return "SI"

# Aplicar la lógica
df_trafos["En_Servicio"] = df_trafos.apply(identificar_reserva, axis=1)

# 3. Guardar el resultado
df_trafos.to_csv(ruta_csv, index=False)

print(f"Archivo actualizado: {ruta_csv}")
# Mostrar resumen
print(df_trafos["En_Servicio"].value_counts())
```
